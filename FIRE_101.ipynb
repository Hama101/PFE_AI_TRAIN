```python
from PIL import Image

def to_jpg(img , i):
    img = Image.open(img)
    rgb_img = img.convert('RGB')
    print("hey")
    rgb_img.save(f'{i}.jpg')
    print(rgb_img)
```

```python
import os, sys

my_list = os.listdir()

i = 1
for file in my_list:
    if  file.endswith('.png'):
        try:
            to_jpg(file , i)
            print("i mm here !")
            i += 1
        except:
            print("nah")
            pass
    
```

```python
import pandas as pd
df = pd.read_csv('https://query.data.world/s/yd2fliikvurreinwa4ovqk5pmkb6sn')
df.sample(10)
```

```python
months = ['Jan' ,'Feb', 'Mar' ,'Apr' ,'May',  'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
data = list(df['Month'])
occ = [data.count(month) for month in months]
print(occ)
csv = {
    "Month" : months,
    "Occ" : occ
}
df_months = pd.DataFrame(csv)
df_months
```

```python
import matplotlib.pyplot as plt

# create a figure and axis
fig, ax = plt.subplots(figsize=(10,10))

# scatter the sepal_length against the sepal_width
ax.scatter(df_months['Month'], df_months['Occ'])
# set a title and labels
ax.set_title('data')
ax.set_xlabel('Month')
ax.set_ylabel('Occ')
```
